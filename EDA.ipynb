```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import read_mouse_data
```

# Diabetic Mouse Data (Full Dataframe)

```python
diabetic_df = read_mouse_data.read_mouse_data('/Users/CHRISTIAN/Downloads/MA591/Diabetic_wound_lactate.xlsx')
diabetic_df
```

# Diabetic Mouse Data (Lactate Only)

```python
diabetic_df_lactate = diabetic_df.iloc[:, 1::2]
# Add a column for the average lactate level at each time point for the diabetic mice, ignoring 0 values
diabetic_df_lactate['average_lactate'] = diabetic_df_lactate.replace(0, np.nan).mean(axis=1, skipna=True)
# Add a column for the standard deviation of lactate level at each time point for the diabetic mice, ignoring 0 values
diabetic_df_lactate['sd_lactate'] = diabetic_df_lactate.replace(0, np.nan).std(axis=1, skipna=True)
diabetic_df_lactate
```

# Diabetic Mouse Data (Wound Closure Only)

```python
diabetic_df_woundclosure = diabetic_df.iloc[:, 2::2]
# Add a column for the average wound closure % at each time point for the diabetic mice
diabetic_df_woundclosure['average_wc'] = diabetic_df_woundclosure.mean(axis=1, skipna=True)
# Add a column for the standard deviation for wound closure % at each time point for the diabetic mice
diabetic_df_woundclosure['sd_wc'] = diabetic_df_woundclosure.std(axis=1, skipna=True)
diabetic_df_woundclosure_reduced = diabetic_df_woundclosure.dropna()
diabetic_df_woundclosure_reduced['time'] = [1, 5, 7, 10]
diabetic_df_woundclosure_reduced
```

# Diabetic Mouse Data (Wound Closure Only INTERPOLATED)

```python
diabetic_df_woundclosure = diabetic_df.iloc[:, 2::2]
diabetic_df_woundclosure = diabetic_df_woundclosure.apply(pd.to_numeric, errors='coerce')
# linear interpolation of wound closure data
diabetic_df_woundclosure_interpolate = diabetic_df_woundclosure.interpolate(method='linear')

diabetic_df_woundclosure_interpolate['average_wc'] = diabetic_df_woundclosure_interpolate.mean(axis=1, skipna=True)

diabetic_df_woundclosure_interpolate['sd_wc'] = diabetic_df_woundclosure_interpolate.std(axis=1, skipna=True)

diabetic_df_woundclosure_interpolate
```

# Healthy Mouse Data (Full Dataframe)

```python
healthy_df = read_mouse_data.read_mouse_data('/Users/CHRISTIAN/Downloads/MA591/Healthy_wound_lactate.xlsx')
healthy_df
```

# Healthy Mouse Data (Lactate Only)

```python
healthy_df_lactate = healthy_df.iloc[:, 1::2]
# Add a column for the average lactate level at each time point for the healthy mice, ignoring 0 values
healthy_df_lactate['average_lactate'] = healthy_df_lactate.replace(0, np.nan).mean(axis=1, skipna=True)
# Add a column for the standard deviation of lactate level at each time point for the healthy mice, ignoring 0 values
healthy_df_lactate['sd_lactate'] = healthy_df_lactate.replace(0, np.nan).std(axis=1, skipna=True)
healthy_df_lactate
```

# Healthy Mouse Data (Wound Closure Only)

```python
healthy_df_woundclosure = healthy_df.iloc[:, 2::2]
# Add a column for the average wound closure % at each time point for the healthy mice
healthy_df_woundclosure['average_wc'] = healthy_df_woundclosure.mean(axis=1, skipna=True)
# Add a column for the standard deviation for wound closure % at each time point for the healthy mice
healthy_df_woundclosure['sd_wc'] = healthy_df_woundclosure.std(axis=1, skipna=True)
healthy_df_woundclosure_reduced = healthy_df_woundclosure.dropna()
healthy_df_woundclosure_reduced['time'] = [1, 5, 7, 10]
healthy_df_woundclosure_reduced
```

# Healthy Mouse Data (Wound Closure Only INTERPOLATED)

```python
healthy_df_woundclosure = healthy_df.iloc[:, 2::2]
healthy_df_woundclosure = healthy_df_woundclosure.apply(pd.to_numeric, errors='coerce')
# linear interpolation of wound closure data
healthy_df_woundclosure_interpolate = healthy_df_woundclosure.interpolate(method='linear')

healthy_df_woundclosure_interpolate['average_wc'] = healthy_df_woundclosure_interpolate.mean(axis=1, skipna=True)

healthy_df_woundclosure_interpolate['sd_wc'] = healthy_df_woundclosure_interpolate.std(axis=1, skipna=True)

healthy_df_woundclosure_interpolate
```

# Individual Diabetic Mouse Plots

```python
# Calculate the max for lactate (this comes from the diabetic data)
min_lactate = 0
max_lactate = diabetic_df[[f'mouse{i}_l' for i in range(1, 11)]].max().max()

# Loop through the range of mouse numbers from 1 to 10
for i in range(1, 11):
    # Create a figure and a set of axes for each plot
    fig, ax1 = plt.subplots()

    # Plot the first axis (lactate)
    ax1.plot(diabetic_df['time'], diabetic_df[f'mouse{i}_l'], color='b')
    ax1.set_xlabel('Time')  # Set x-axis label
    ax1.set_ylabel(f'Diabetic Mouse {i} Lactate', color='b')  # Set y-axis label for lactate
    ax1.tick_params(axis='y', labelcolor='b')  # Set color for the first y-axis ticks
    ax1.set_ylim(min_lactate, max_lactate)  # Set constant y-axis limits for lactate

    # Create the second axis (wound closure percentage)
    ax2 = ax1.twinx()
    ax2.plot(diabetic_df['time'], diabetic_df_woundclosure_interpolate[f'mouse{i}_wc'], color='r')
    ax2.set_ylabel(f'Diabetic Mouse {i} Wound Closure %', color='r')  # Set y-axis label for wound closure
    ax2.tick_params(axis='y', labelcolor='r')  # Set color for the second y-axis ticks
    ax2.set_ylim(0, 100)  # Set constant y-axis limits for wound closure

    # Add dashed green lines at y = 5 and y = 15
    ax1.axhline(y=5, color='g', linestyle='--')
    ax1.axhline(y=15, color='g', linestyle='--')

    # Add titles and legends
    ax1.set_title(f'Diabetic Mouse {i} Lactate and Wound Closure Over Time')

    plt.show()
```

# Average of Diabetic Mice Plot

```python
# Create a figure and a set of axes for each plot
fig, ax1 = plt.subplots()

# Plot the first axis (lactate)
ax1.plot(healthy_df['time'], healthy_df_lactate['average_lactate'], color='b')
ax1.plot(healthy_df['time'], healthy_df_lactate['average_lactate'] + healthy_df_lactate['sd_lactate'], color='b', linestyle='--') # + sd
ax1.plot(healthy_df['time'], healthy_df_lactate['average_lactate'] - healthy_df_lactate['sd_lactate'], color='b', linestyle='--') # - sd
ax1.set_xlabel('Time')  # Set x-axis label
ax1.set_ylabel('Average Lactate', color='b')  # Set y-axis label for lactate
ax1.tick_params(axis='y', labelcolor='b')  # Set color for the first y-axis ticks
ax1.set_ylim(min_lactate, max_lactate)  # Set constant y-axis limits for lactate

# Create the second axis (wound closure percentage)
ax2 = ax1.twinx()
ax2.plot(healthy_df_woundclosure_reduced['time'], healthy_df_woundclosure_reduced['average_wc'], color='r')
ax2.plot(healthy_df_woundclosure_reduced['time'], healthy_df_woundclosure_reduced['average_wc'] + healthy_df_woundclosure_reduced['sd_wc'], color='r', linestyle='--') # + sd
ax2.plot(healthy_df_woundclosure_reduced['time'], healthy_df_woundclosure_reduced['average_wc'] - healthy_df_woundclosure_reduced['sd_wc'], color='r', linestyle='--') # - sd
ax2.set_ylabel('Average Wound Closure %', color='r')  # Set y-axis label for wound closure
ax2.tick_params(axis='y', labelcolor='r')  # Set color for the secondy-axis ticks
ax2.set_ylim(0, 100)  # Set constant y-axis limits for woundclosure

# Add dashed green lines at y = 5 and y = 15
ax1.axhline(y=5, color='g', linestyle='--')
ax1.axhline(y=15, color='g', linestyle='--')

# Add titles and legends
ax1.set_title('Healthy Mouse Average Lactate and Wound Closure Over Time')

# Show the plot
plt.show()
```
